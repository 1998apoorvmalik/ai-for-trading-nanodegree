```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set plotting options
%matplotlib inline
plt.rc('figure', figsize=(16, 9))
```

```python
# use returns to create a price series
drift = 100
r1 = np.random.normal(0, 1, 10000) 
s1 = pd.Series(np.cumsum(r1), name='s1') + drift
s1_lagged = s1.shift(-1).dropna()
s1 = s1[:-1]

# s1, s1_lagged = s1_lagged, s1
s1.plot(figsize=(20,10))
plt.show()
```

```python
np.mean(s1), np.std(s1)**2
```

```python
beta = 0.999
eps = np.random.normal(0, 3, len(s1_lagged))
s1_hat = beta*s1_lagged + eps

s1_hat.plot(figsize=(20,10))
```
